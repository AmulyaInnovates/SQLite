```python
import sqlite3
database= "Analyzing_2.sqlite"
conn= sqlite3.connect(database)
print("Connection was successfull")
```

```python
import pandas as pd
tables= pd.read_sql("""SELECT * FROM sqlite_master WHERE type='table' ;""", conn)
tables
```
